# VGG16 - Pretrained Model

```python
from tensorflow.keras.applications import vgg16

vgg16_model = vgg16.VGG16(weights='imagenet')
vgg16_model.summary()
```

```python
# Utility Function to Load Image, Preprocess input and Targets
def predict_image(model, img_path, preprocess_input_fn, decode_predictions_fn, target_size=(224, 224)):

    img = image.load_img(img_path, target_size=target_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input_fn(x)
    
    preds = model.predict(x)
    predictions_df = pd.DataFrame(decode_predictions_fn(preds, top=10)[0])
    predictions_df.columns = ["Predicted Class", "Name", "Probability"]
    return predictions_df
```

## Single Image of Komodo Dragon

```python
#img_path="rocking_chair.png"  ## Uncomment this and put the path to your file here if desired
# Predict Results
predict_image(vgg16_model, single_1_img, vgg16.preprocess_input, vgg16.decode_predictions)
```

1. Yes, the model got the correct answer.
2. The model was 99% confident
3. This image only contianed a single object, the rest of the image was background grass.
4. All the other possible classes listed were animals the resembles reptiles or lizzards.  Since the classifier can't intereprite the actual size of the animal it makes sense how it would look like a lizard given the shape.
